```python
import re
def clean_tweet(tweet):
  #pattern that should remove any emoticon that is present in the training set
  pattern = "(\:\w+\:|\<[\/\\]?3|[\(\)\\\D|\*\$][\-\^]?[\:\;\=]|[\:\;\=B8][\-\^]?[3DOPp\@\$\*\\\)\(\/\|])(?=\s|[\!\.\?]|$)"
  #patter that will remove hyperlinks from the training set
  pattern2 = "(https?:\/\/(?:www\.|(?!www))[a-zA-Z0-9][a-zA-Z0-9-]+[a-zA-Z0-9]\.[^\s]{2,}|www\.[a-zA-Z0-9][a-zA-Z0-9-]+[a-zA-Z0-9]\.[^\s]{2,}|https?:\/\/(?:www\.|(?!www))[a-zA-Z0-9]+\.[^\s]{2,}|www\.[a-zA-Z0-9]+\.[^\s]{2,})"

  row_str = tweet
  row_str = re.sub(pattern, " ", row_str)
  row_str = re.sub(pattern2, " ", row_str)

  #this is aimed at removing any recurring characters of more than length 2 to just recurring characters of length 2
  row_str = re.sub("([a-zA-Z!.])\\1\\1+", r"\1\1", row_str);

  #this is aimed at removing any username tagged in the tweet
  row_str = re.sub('@[^\s]+','',row_str)

  #this is aimed at replacing '@' chat lingo to its equivalent meaning 'at'
  row_str = row_str.replace('@', 'at')

  #this is aimed at replacing any hashtags mentioned in the tweet. NOTE: Only the hash symbol is removed but not the tag associated wit it.
  row_str = row_str.replace('#', ' ')

  return row_str
```
